```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
from sklearn.cluster import KMeans
import numpy as np

# Generate sample geographic points
data = {
    'longitude': [-0.1276, -0.1277, -0.1278, -0.1279, -0.1280, -0.1281, -0.1282],
    'latitude': [51.5074, 51.5075, 51.5076, 51.5077, 51.5078, 51.5079, 51.5080]
}

df = pd.DataFrame(data)
geometry = [Point(xy) for xy in zip(df.longitude, df.latitude)]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Train a KMeans model
kmeans = KMeans(n_clusters=2, random_state=0).fit(df[['longitude', 'latitude']].values)

# Ask for new geographic points
while True:
    try:
        longitude = float(input("Enter longitude: "))
        latitude = float(input("Enter latitude: "))
        activity_name = input("Enter activity name: ")

        # Predict the cluster for the new point
        cluster = kmeans.predict([[longitude, latitude]])[0]

        # Display the map with the new point
        fig, ax = plt.subplots(1, 1, figsize=(8.27, 11.69))  # A4 size in inches
        world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
        base = world.plot(ax=ax, color='white', edgecolor='black')
        gdf.plot(ax=base, marker='o', color='red', markersize=50)
        ax.scatter(longitude, latitude, color='blue', s=100, label=activity_name)
        plt.legend()
        plt.show()

        # Ask if the user wants to add another point
        add_another = input("Do you want to add another point? (y/n): ")
        if add_another.lower() != 'y':
            break

    except ValueError:
        print("Please enter valid longitude and latitude values.")
```
